## Opening text file

```python
file = open('sample.txt','r')
```

```python
content = file.read()
content
```

### Tokenization

```python
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
```

```python
sentences = sent_tokenize(content)
```

```python
sentences
```

```python
words = []
tokenizer = RegexpTokenizer("[\w']+")
for sentence in sentences:
    words.extend(tokenizer.tokenize(sentence))
```

```python
words
```

### Removing stop words and POS tagging

```python
from nltk.corpus import stopwords
stopWords = stopwords.words('english')
```

#### Extracting words without stop words

#### POS tagging words after separating stop words

```python
wordsWithoutStopWords = []
tokenizer = RegexpTokenizer("[\w']+")
for sentence in sentences:
    w = tokenizer.tokenize(sentence)
    filteredWords = [i for i in w if i.lower() not in stopWords]
    wordsWithoutStopWords.extend(filteredWords)
    print(f"Sentence:{w}")
    print(f"Sentence without stopwords:{filteredWords}")
    print(f"POS tagging:{nltk.pos_tag(filteredWords)}")
    print()
```

### Stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
for word in words:
    print(f"{word} : {stemmer.stem(word)}")
```

### Lemmatization

```python
from nltk import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for word in words:
    print(f"{word}:{lemmatizer.lemmatize(word)}")
```

```python
hashMap = {}
for word in words:
    hashMap[word] = (stemmer.stem(word),lemmatizer.lemmatize(word))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(hashMap,orient='index')
df.rename(columns={0:"Stemming",1:'Lemmatization'},inplace=True)
```

### Creating a corpus (Using three sentences of text file)

```python
corpus = sentences[:3]
corpus
```

#### Creating a set of unique words in document

```python
wordCorpus = []
for sentence in corpus:
    wordCorpus.extend(tokenizer.tokenize(sentence))
```

```python
wordSet = set(wordCorpus)
print(f"Number of unique words: {len(wordSet)}")
print(f"Number of words : {len(wordCorpus)}")
```

### Computing Term Frequency

```python
import numpy as np
```

```python
termFrequency = pd.DataFrame(np.zeros((len(corpus), len(wordSet))), columns=wordSet)
```

```python
print("Frequency of words in each sentence")
for i,sentence in enumerate(corpus):
    w = tokenizer.tokenize(sentence)
    wordFrequency = {l:0 for l in w}
    for j in w:
        termFrequency[j][i] = termFrequency[j][i] + (1/len(w))
        wordFrequency[j] += 1
    print(f"{sentence}\n{wordFrequency}\n")
```

```python
termFrequency
```

### Computing Inverse Document Frequency

```python
idf = {}
for j in wordSet:
    k = 0
    for i,sentence in enumerate(corpus):
        w = tokenizer.tokenize(sentence)
        if j in w:
            k += 1
    idf[j] = np.log10(len(corpus) / k)
    print(f"{j}: {idf[j]}")
```

### Computing Term Frequency-Inverse Document Frequency

```python
tf_idf = termFrequency.copy()
```

```python
for j in wordSet:
    for i in range(len(corpus)):
        tf_idf[j][i] = termFrequency[j][i] * idf[j]
```
