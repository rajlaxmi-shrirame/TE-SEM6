# Handling NaN Values
### Replacing it by median values

```python
df["Car"].fillna(df["Car"].median(skipna=True),inplace=True)
df["BuildingArea"].fillna(df["BuildingArea"].median(skipna=True),inplace=True)
df["YearBuilt"].fillna(df["YearBuilt"].median(skipna=True),inplace=True)
```

### Droping rows with NaN values in Council Area column

```python
df = df.dropna(subset=["CouncilArea"])
```

```python
pd.to_datetime(df["Date"],format="%d/%m/%Y")
```

```python
df = df.astype({'YearBuilt':'int','Postcode':'int','Bedroom2':'int','Bathroom':'int'})
```
